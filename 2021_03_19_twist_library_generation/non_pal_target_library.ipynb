#### Shortened base editing targets

Our project will test if shortened guides alter the base editing pattern of CRISPR targets. 
We'll generate a number of diverse target sequences to test this with A/Ts (and C/Gs) 
spaced at sites known to edit effectively. Our general patterns will be:

guide - scaffold - spacer - target

For the target the specific patterns will be:

  --- -----
NNNNN NNNNN NNNNN NNNN NGG

NNANA NANAN NNNNN NNNN NGG
NANAN ANANA NNNNN NNNN NGG


```python
import random
def random_sequence(length):
    return("".join(random.choices(['A','C','G','T'],k=length)))

def gc(seq):
    return((sum([1 if x == 'C' or x == 'G' else 0 for x in seq]))/len(seq))

def screen_gc(minval,maxval):
    score = -1
    newseq = ""
    while score < minval or score > maxval:
        newseq = random_sequence(20)
        score = gc(newseq)
    return(score,newseq)

for i in range(0,10):
    print(screen_gc(0.4,0.6))
```

```python
known_targets = {}
known_targets_no_pam = {}

def create_random_target_2mer(frontpad,tuples_to_choose=3):
    pams = ['AGG','CGG','TGG']
    tuples = ['TA','CA','GA']
    chosen = random.choices(tuples,k=tuples_to_choose)
    return(random_sequence(frontpad) + 
           "".join(chosen) + 
           random_sequence(20 - frontpad - len("".join(chosen))) +
           "".join(random.sample(pams,k=1)))

def create_random_target(frontpad):
    pams = ['AGG','CGG','TGG']
    tuples = ['TAT','TAC','TAG','TAA','CAT','CAC','GAT']
    chosen = random.sample(tuples,k=3)
    return(random_sequence(frontpad) + 
           "".join(chosen) + 
           random_sequence(20 - frontpad - (9)) +
           "".join(random.sample(pams,k=1)))

for i in range(0,500):
    tgt = create_random_target(1)
    subset = tgt[0:20]
    if not subset in known_targets_no_pam:
        known_targets_no_pam[subset] = tgt
        known_targets[tgt] = "front_pad1_triples"
        
for i in range(0,500):
    tgt = create_random_target(0)
    subset = tgt[0:20]
    if not subset in known_targets_no_pam:
        known_targets_no_pam[subset] = tgt
        known_targets[tgt] = "front_pad0_triples"

for i in range(0,1000):
    tgt = create_random_target_2mer(0,4)
    subset = tgt[0:20]
    if not subset in known_targets_no_pam:
        known_targets_no_pam[subset] = tgt
        known_targets[tgt] = "front_pad0_doubles"
```
